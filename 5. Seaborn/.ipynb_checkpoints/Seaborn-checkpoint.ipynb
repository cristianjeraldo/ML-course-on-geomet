# Seaborn

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

```python
geomet = pd.read_csv('abrasion_res.csv')
```

```python
geomet.head()
```

```python
geomet.columns
```

## Distribution Plots

```python
sns.histplot(geomet['Ai'])
```

```python
# Changing number of bins
sns.histplot(geomet['Ai'], kde=False, bins=30)
```

```python
sns.jointplot(x='Ai',y='K-Feldspar',data=geomet)
```

```python
sns.jointplot(x='Ai',y='K-Feldspar',data=geomet,kind='hex')
```

```python
sns.jointplot(x='Ai',y='K-Feldspar',data=geomet,kind='reg')
```

```python
sns.jointplot(x='Ai',y='K-Feldspar',data=geomet,kind='kde')
```

```python
sns.pairplot(geomet)
```

```python
sns.pairplot(geomet,hue='minzone')
```

```python
sns.pairplot(geomet,hue='minzone',palette='magma')
```

```python
sns.pairplot(geomet,hue='minzone',palette=['blue','orange','purple'],)
```

## Categorical Plots

```python
sns.barplot(x='minzone',y='Ai',data=geomet)
```

```python
sns.countplot(x='minzone',data=geomet)
```

```python
sns.boxplot(x='minzone',y='Ai',data=geomet)
```

```python
sns.boxplot(x='minzone',y='Ai',data=geomet,hue='alteration')
```

## Matrix Plots

```python
geomet.corr()
```

```python
gc = geomet.corr()
```

```python
sns.heatmap(gc)
```

```python
# Annotations
sns.heatmap(gc,annot=True)
```

```python
# Color maps
sns.heatmap(gc,annot=True, cmap='coolwarm')
```

```python
# FacetGrid
g = sns.FacetGrid(data=geomet,col='minzone',row='alteration')
```

```python
g = sns.FacetGrid(data=geomet,col='minzone',row='alteration')

g.map(sns.histplot,'Ai')
```

```python
g = sns.FacetGrid(data=geomet,col='minzone',row='alteration')

g.map(sns.scatterplot,'Ai','K-Feldspar')
```

## Regression Plots

```python
sns.lmplot(x='K-Feldspar',y='Ai',data=geomet)
```

```python
sns.lmplot(x='K-Feldspar',y='Ai',data=geomet,hue='minzone')
```

```python
sns.lmplot(x='K-Feldspar',y='Ai',data=geomet,col='minzone')
```

```python
sns.lmplot(x='K-Feldspar',y='Ai',data=geomet,col='minzone',row='alteration')
```
