# Creating New Column GMU

```python
# We will generate a new column GMU containing the 6 defined GMUs for Ai
```

```python
conditions = [
    (df['alteration'] == 'AR') & (df['minzone'] != 'PRI'),
    (df['alteration'] == 'CLO-SER') & (df['minzone'] != 'PRI'),
    (df['alteration'] != 'POT') & (df['minzone'] == 'PRI'),
    (df['alteration'] == 'POT') & (df['minzone'] == 'SSF'),
    (df['alteration'] == 'POT') & (df['minzone'] == 'SSD'),
    (df['alteration'] == 'POT') & (df['minzone'] == 'PRI')
]
```

```python
values = ['GMU1', 'GMU2', 'GMU3', 'GMU4','GMU5','GMU6']
```

```python
df['GMU'] =np.select(conditions, values)
```

```python
df
```

```python
palette_gmu = ['orange', 'blue','black','green','red','purple']
```

```python
fg = (
    sns.FacetGrid(data=df, hue='GMU', aspect=2, height=5, palette=palette_gmu)
        .map(probscale.probplot, 'Ai', plottype='prob', probax='y')
        .set_axis_labels(x_var='Ai', y_var='Probability')
        .add_legend()
)
```

```python
df.groupby('GMU')['Ai'].count()
```

```python
df.groupby('GMU')['Ai'].mean().round(2)
```

```python
fig,ax = plt.subplots(figsize=(6,4))
sns.boxplot(data=df,x='GMU',y='Ai',palette=palette_gmu);
```
