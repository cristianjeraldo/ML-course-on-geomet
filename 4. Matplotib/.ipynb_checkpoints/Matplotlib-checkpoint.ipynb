```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

```python
# Object Oriented (OO) interface. Instantiate figure objects and then call method from that object

fig,ax = plt.subplots()
```

```python
# fig,ax = plt.subplots()

# axes.plot(x,y)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2)

# Fix overlapping, plt.tight_layout()
```

```python
# Also is possible to index
fig,ax = plt.subplots(nrows=1,ncols=2)

ax[0].plot(x,y)
ax[1].plot(y,x);
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2)

ax[0].plot(x,y)
ax[0].set_title('Plot 1')

ax[1].plot(y,x)
ax[1].set_title('Plot 2');
```

```python
# figsize and dpi
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(8,2))

ax[0].plot(x,y)
ax[0].set_title('Plot 1')

ax[1].plot(y,x)
ax[1].set_title('Plot 2');
```

```python
# to save a figure
fig.savefig('my_fig.png',dpi=200)
```

```python
fig,ax = plt.subplots()

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc=9);
```

```python
# Color
fig,ax = plt.subplots()

ax.plot(x,x**2,label='X Squared',color='blue')
ax.plot(x,x**3,label='X Cubed',color='red')

ax.legend(loc=9);
```

```python
# Line width
fig,ax = plt.subplots()

ax.plot(x,x**2,label='X Squared',color='blue',linewidth='5')
ax.plot(x,x**3,label='X Cubed',color='red',linewidth='2')

ax.legend(loc=9);
```
